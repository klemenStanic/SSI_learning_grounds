## 1. Ecosystem overview

### 1.1 Self-sovereign identity (SSI)
Sources: 
- https://sovrin.org/wp-content/uploads/2017/06/The-Inevitable-Rise-of-Self-Sovereign-Identity.pdf
- https://en.wikipedia.org/wiki/Self-sovereign_identity

To create the long-missing identity layer of the Internet, a new, trusted infrastructure is required which enables identity owners to share not only identity, but also verified attributes about people, organisations and things.

For identities to be truly self-sovereign, this infrastructure needs to reside in an environment of diffuse trust, not belonging to or controlled by any single organisation or even a small
group of organisations. *Distributed ledger technology (DLT)* [Sunyaev, Ali. "Distributed ledger technology." Internet Computing. Springer, Cham, 2020. 265-299.]
is the breakthrough that makes this possible. It enables multiple institutions, organisations and governments to work together for the first time by forming a decentralised network much like the Internet itself, where data is replicated in multiple locations to be resistant to faults and tampering. While distributed ledger technology has been around for some time, new DLTs such as Bitcoin[Nakamoto, S. (2008) Bitcoin: A Peer-to-Peer Electronic Cash System. https://bitcoin.org/bitcoin.pdf] and Ethereum[Buterin V. Ethereum white paper. GitHub repository. 2013;1:22-3.] have resulted in a greater realisation of its potential,
particularly with respect to decentralisation and security.

When combined with distributed key management and peer-to-peer sharing of encrypted claims, DLT is what finally makes self-sovereign identity possible. Within this identity layer, mechanisms for discovery, routing of requests, exchanging of data and recording events can exist pervasively, with no single entity being in control.

The three pillars of the eco-system are: *Decentralized identifiers (DIDs)*[https://www.w3.org/TR/did-core], *Verifiable Credentials (VCs)*[https://www.w3.org/TR/vc-data-model/] and the *Verifiable data registry (Blockchain)*[https://www.w3.org/TR/vc-data-model/#dfn-verifiable-data-registries].

### 1.2 Decentralized Identifiers (DIDs)
Sources: 
- https://www.w3.org/TR/did-core/
- https://w3c-ccg.github.io/did-primer/

For self-sovereign identity, which can be defined as a lifetime portable digital identity that does not depend on any centralized authority, we need a new class of identifier that fulfills four requirements: persistence, global resolvability, cryptographic verifiability, and decentralization.

To achieve this, the decentralized identifier (DID) was proposed. At a superficial level, a DID is simply an URI [Uniform Resource Identifier (URI): Generic Syntax. T. Berners-Lee; R. Fielding; L. Masinter. IETF. January 2005. Internet Standard. URL: https://www.rfc-editor.org/rfc/rfc3986], which represents a new type of a globally unique identifier. At a deeper level, DIDs are the core component of an entirely new layer of decentralized digital identity and public key infrastructure for the Internet. 

#### 1.2.1 DID arhitecture

Overview of the DID architecture and the relationship between basic components.
Source: Based on W3C DID.

Figure [TODO ref] shows an overview of the DID architecture. A *DID subject* is the entity, identified by the DID and can be a person, an organisation, a thing or a concept. A *DID controller* is the entity (person, organisation or anutonomous software) that has the capability to make changes to the DID document. Note that the DID Subject can also be the DID Controller.  A *DID URL* extends the syntax of a basic DID to incorporate other standard URI components such as path, query and fragment in order to locate a particular resource; for example, a cryptographic public key inside a DID document, or a resource external to the DID document.
As shown on the figure, a DID refers to a specific DID subject and resolves to a DID document, which is controlled by the DID controller. Both the DID and the DID document are recorded on the Verifiable Data Registry. A DID URL refers and dereferences to a DID document.


#### 1.2.2 DID format

</div>
DID format specification.

Figure [TODO ref] shows that a DID string is composed of three parts: a *Scheme*, a *DID Method* and a *DID Method Specific String*.
- The Scheme defines that the DID string represents a Decentralized Identifier. 
- The DID Method defines the type of a ledger, that the DID is associated with.  For example, if the Sovrin Public Ledger[https://sovrin.org/] is used, the DID Method is *sov*. Similarily, if the Bitcoin blockchain is used, the DID Method is *btcr*.
- The DID method specific string represents a string, which is unique in the namespace of a specific ledger, defined by the DID method. The format of the DID Method Specific String is defined by the DID method and is different for every type of ledger used. 


#### 1.2.3 DID documents
The DID architecture can be thought of as a global key-value database, where the key is a DID, and the value corresponding to the DID is a *DID document*. The process of obtaining the DID Document from the DID is called resolving the DID. The purpose of the DID document is to describe the public keys, authentication protocols and service endpoints necessary to bootstrap cryptographically-verifiable interactions with the identified entity. The DID Document uses the JavaScript Object Notation for Linked Data (JSON-LD) [https://www.w3.org/TR/json-ld11/] format, which makes it human-readable and easy to use and store in unstructured databases. 

```
{
  "@context": [
    "https://www.w3.org/ns/did/v1",
    "https://w3id.org/security/suites/ed25519-2020/v1"
  ]
  "id": "did:example:123456789abcdefghi",
  "authentication": [{
    
    "id": "did:example:123456789abcdefghi#keys-1",
    "type": "Ed25519VerificationKey2020",
    "controller": "did:example:123456789abcdefghi",
    "publicKeyMultibase": "zH3C2AVvLMv6gmMNam3uVAjZpfkcJCwDwnZn6z3wXmqPV"
  }]
}
```

### 1.3 Verifiable Credentials (VCs)
Sources: 
- https://www.w3.org/TR/vc-data-model
- https://en.wikipedia.org/wiki/Verifiable_credentials
- https://w3c.github.io/vc-use-cases

It is currently difficult to transmit credentials such as driver's licenses, proofs of age, education qualifications, and healthcare data, via the Internet in a way that is verifiable yet protects individual privacy.

Verifiable Credentials are useful when a person needs to prove that they are:

- above a certain age,
- capable of driving a particular motor vehicle,
- require a particular medication,
- trained and certified as an electrician,
- professionally licensed to practice medicine,
- cleared to travel internationally *etc*.

Note that these use-cases are merely high-level examples. More use-cases, described more in-depth, can be found at [W3C VC-use-cases](https://w3c.github.io/vc-use-cases).

#### 1.3.1 Roles
The Verifiable Credentials ecosystem is composed of four primary roles, presented in the figure [TODO ref]. Note that the figure provides only an example ecosystem, and other variations exist.


Roles and information flows in the VC ecosystem.
Source: Figure based on W3C vc-data-model

1) The *Issuer* asserts claims about one or more subjects, creates verifiable credentials from these claims and transmits the verifiable credentials to the Holder. Issuers can be corporations, non-profit organizations, governments, individuals, *etc*.

2) The *Holder* stores verifiable credentials on behalf of a Subject. Holders are typically also the Subjects of a credential. When requested, a Holder generates and transmits a Verifiable Presentation to a Verifier. Holders can be students, employees, customers, *etc*.

3) The *Verifier* requests a *Profile/Presentation* of the Subject. A Profile/Presentation contains a specific set of credentials. The Verifier verifies that the credentials provided in the Profile/Presentation are fit-for-purpose. Example Verifiers include employers, security personel and websites.

4) The *Verifiable Data Registry* is a mechanism, used to store identifiers, keys and other relevent data, such as verifiable credential shemas, revocation registries, issuer public keys *etc*. A Verifiable Data Registry can be a trusted database, a decentralized database, a government ID database or a distributed ledger.  


#### 1.3.2 Core data model concepts

To establish trust using a Verifiable credential, data has to be exchanged between the different roles in the VC ecosystem. The data that is exchanged differs based on the roles participating, but is fundamentally composed of *Claims*, *Credentials*, and *Profiles/Presentations*. 

- A *Claim* is a statement about a subject, expressed as a subject-property-value relationship.
This enables a wide variety of statements, e.g. whether or not someone is over the age of 21, or whether someone graduated from a particular university. Such a claim is presented in the *Example 1* in the figure [TODO ref].
These claims may be merged together to express a graph of information about a particular subject. The *Example 2* extends the data model by adding claims that state that Alice knows Bob and that Bob is a professor.

Basic structure of a claim and merging of claims into information graphs. Source: based on W3C vc-data-model

- A *Verifiable Credential(VC)*, whose structure is shown in the figure [TODO ref], is a set of multiple claims, combined into a single document, where the data is digitally signed (Proof). Some basic information about the Verifiable Credential, called Credential Metadata, is added. A Verifiable Credential is usually sent from the Issuer to the Holder.

### 1.4 Verifiable data registry
Sources: 
- https://www.w3.org/TR/vc-data-model/
- https://en.wikipedia.org/wiki/Distributed_ledger

A *Verifiable Data Registry* is a role a system performs by mediation between the creation and verification of identifiers, keys and other relevant data, such as verifiable credential schemas, revocation registries, issuer public keys *etc.*, which are required to use verifiable credentials. 
The most commonly used verifiable data registry is a distributed ledger, built atop of the Distributed Ledger Technology(DLT)[Rauchs, Michel, et al. "Distributed ledger technology systems: A conceptual framework." Available at SSRN 3230013 (2018)], commonly called Blockchain Technology. This approach is used in order to establish a decentralized database, which provides control over the evolution of data between entities through a peer-to-peer network. Multiple *consensus algorithms*[https://en.wikipedia.org/wiki/Consensus_(computer_science)] are used to ensure replication of data across the nodes of the network.
In other words, the distributed ledger or a blockchain enables everyone in the network to have the same source of thruth about which credentials are valid and who attested to the validity of the data inside the credential, without revealing the actual data. 

The main types of data, that are written to the blockchain in the context of Verifiable Credentials, include:
- Public DIDs
- Credential Definitions: 
Credential definitions represent the templates to be stored on the ledger for different (often tangible) proofs of identity or qualification issued by authorities. These include drivers licenses, passports, identification cards, credit cards, *etc*.
- Schemas: 
Schemas provide formal descriptions for the structure of a credential or a template, which outlines the verified data you can issue or verify from the users.

#### 1.5.1 Wallets, Private and Public DIDs
Source: https://identity.foundation/peer-did-method-spec/

Aries allows for the usage of the *peer* DID method using private DIDs. We demonstrate the concept on an example: 

In traditional DID methods, a public entity (e.g. a *University*) would register its DID on the public ledger, allowing everyone to access it and resolve it. When a separate entity (e.g. *Bob*, a student at the University) wants to establish a connection to the University, he resolves the University's DID, creates a pairwise DID and publishes it to the public ledger. Similarily, the *University* creates a pairwise DID and also publishes it to the public ledger. With both entities having a pairwise DID each, a connection can be established. This poses a couple of problems: for every student at the *University*, a separate pairwise DID must be created and published to the public ledger, which creates a big number of ledger transactions, decresing the performance and scalability of the whole system. Furthermore, everyone can see and resolve the DIDs of the *Bob-University* relationship. 

To circumvent this, we use a *peer* DID method, which uses *private/pairwise DIDs*. The procedure is presented in the diagram [TODO ref]. Similar to the first example, the University creates and publishes the DID and a DID document to the public ledger. Bob resolves the DID and creates a pairwise DID and a DID document. He packages the DID and the DID document and sends it to the University. The University stores the received data in a data structure called a *wallet*. When necessary, the University can resolve that DID by checking the wallet, not the public ledger. Likewise, the University creates its pairwise DID and DID document for the University-Bob relationship and sends the data to Bob. Bob stores the data in a wallet. 

````python
import iplantuml
````

````python
%%plantuml

@startuml
actor "University" as U
actor "Bob" as B
database "Public ledger" as L

autonumber
U --> U: Create a public DID and a DID document
U --> L: Publish DID and DID document
B --> L: Resolve University's DID
B --> B: Create Bob's pairwise DID and DID document for\nthe Bob:University relationship
B --> U: Send data
U --> U: Store data in wallet
U --> U: Create University's pairwise DID and DID document\nfor the Bob:University relationship
U --> B: Send data
B --> B: Store data in wallet
@enduml
````

### 2.2 Create a DID and publish it to the ledger

Alice, who acts as the Issuer, instantiates her own DID and publishes it on the local von-network ledger. This is done by sending a POST HTTP request to the ledger's endpoint, where the ledger creates and publishes the DID. A seed, the role and alias must be provided alongside the POST request. The seed represents a string, which is used to generate the private and public keys, that are associated with the registered DID. The "TRUST_ANCHOR" role specifies that Alice is a trustworthy entity, that can issue credentials. The alias represents a name, that is associated with the DID. 

Note that the same could be achieved by manually registering the DID using the web user interface, provided by the von-network (accessible on `localhost:9000`).

````python
import requests
import json
ledger_url = "http://localhost:9000"

def send_request(url, data={}, req_type="post"):
    if req_type == "get":
        headers = {'content-type': 'application/json'}
        return requests.get(url, data=json.dumps(data), headers=headers, verify=True, allow_redirects=False)
    
    headers = {'content-type': 'application/json'}
    return requests.post(url, data=json.dumps(data), headers=headers, verify=True, allow_redirects=False)


def register_DID_on_ledger(seed, role, alias):
    d = { 
            "seed": seed,
            "role": role,
            "alias": alias 
    }
    return send_request(ledger_url + "/register", d)



resp = register_DID_on_ledger("Alice000000000000000000000000001", "TRUST_ANCHOR", "Alice")

alice_did = resp.json()["did"]
print(f"Alice's DID: {alice_did}")
print(f"Response: \n{resp.text}")
````

Note that the generated DID, that is received from the ledger, only represents the *DID method specific string*. The *Scheme* and the *DID method* are implicitly set on the ledger.

### 2.3 Issuer(Alice) initialization:
In order to start the Alice agent, using the aca-py Python script, the following command must be ran in a separate terminal.

```
aca-py start \
--label Alice \
-it http 0.0.0.0 8000 \
-ot http \
--admin 0.0.0.0 11000 \
--admin-insecure-mode \
--genesis-url http://localhost:9000/genesis \
--seed Alice000000000000000000000000001 \
--endpoint http://localhost:8000/ \
--debug-connections \
--auto-provision \
--wallet-type indy \
--wallet-name Alice \
--wallet-key secret
```

The command line options define the behavior of the agent: 
- `--label Alice` This is the label or name that is given to the agent instance,
- `-it http 0.0.0.0 8000` and `-ot http` are the inbound and outbound transport methods that ACA-py uses to communicate to other ACA-py instances,
- `--admin 0.0.0.0 11000` and `--admin-insecure-mode` are the parameters that configure how the controller application can communicate with ACA-py. In this case, the Admin Endpoints are available on port 11000, and insecure, meaning there is no authentication required,
- `--genesis-url http://localhost:9000/genesis` This is the URL to the genesis file. When a schema and credential definition are created, transactions are written to the Indy ledger. To be able to create these transactions, ACA-py needs to know about the genesis transaction,
- `--seed Alice000000000000000000000000001` This is the seed value used to register the DID. This seed value proves ownership of the public DID, since the public and private keys are both derived from it,
- `--endpoint http://localhost:8000/` This is the URL that ACA-py will send to ledger, to register where the ACA-py instance for your DID can be reached,
- `--debug-connections` This ensures a more verbose output,
- `--auto-provision` This parameter makes sure that ACA-py is going to create a wallet, when it doesnt exist,
- `--wallet-type indy`, `--wallet-name Alice` and `--wallet-key secret` are the parameters that are used to create the wallet. In this setup, the wallet is stored in files on your system. The wallets can be found in `~/.indy_client/wallet/`. The key is required to write and read to the wallet.

### 2.5 Connecting Alice and Bob
A secure connection between the two entities is established through the sequence of messages, presented in figure [TODO ref]. Alice first registers her DID on the Ledger, which has already been done. Then, she creates an invitation, which she sends to Bob out of band. Bob then receives the invitation, accepts it and sends a connection request to Alice (which happens automatically). Alice finally accepts the connection request.

````python
%%plantuml

@startuml
actor "Alice(issuer)" as A
actor "Bob(holder)" as B
database "Ledger" as L

autonumber
A -> L : Register the DID on the Ledger
A -> A : Create the invitation
A --> B : Alice sends the invitation to Bob, out of band
B -> B : Bob receives the invitation
B -> A: Bob accepts the invitation (Alice automatically receives the connection request)
A -> A: Alice accepts the connection request
@enduml
````

#### 2.5.1 Invitation
Alice generates an invitation by invoking the `/out-of-band/create-invitation` endpoint, where a handshake protocol is specified. The `use_public_did` parameter is set to False. The reason for this is that peer DIDs are used between the Alice and Bob.

````python
def pp_json(json_thing, sort=True, indents=2):
    if type(json_thing) is str:
        out = json.dumps(json.loads(json_thing), sort_keys=sort, indent=indents)
    else:
        out = json.dumps(json_thing, sort_keys=sort, indent=indents)
    return out

alice_endpoint = "http://localhost:11000"
bob_endpoint = "http://localhost:11001"

def create_invitation():
    d = {
        "handshake_protocols": ["did:sov:BzCbsNYhMrjHiqZDTUASHg;spec/didexchange/1.0"],
        "use_public_did": False
    }
    return send_request(alice_endpoint + "/out-of-band/create-invitation", d)

resp = create_invitation()
invitation = resp.json()["invitation"]
print(f"Invitation: \n{pp_json(invitation)}")
````

The invitation can be passed to Bob as a link, or a QR code. A non-repudiable signature could be added to the invitation. Such a solution is proposed in [Aries RFC 0066: Non-Repudiable Signature for Cryptographic Envelope](https://github.com/hyperledger/aries-rfcs/blob/master/features/0066-non-repudiable-cryptographic-envelope/README.md). 

#### 2.5.2 Receive invitation
Bob receives the invitation.

````python
def receive_invitation(invitation):
    return send_request(bob_endpoint + "/out-of-band/receive-invitation", invitation)

resp = receive_invitation(invitation)
       
bob_connection_id = resp.json()["connection_id"]
print(f"Response: \n{pp_json(resp.json())}")
print(f"Bobs connection id: {bob_connection_id}")
````

#### 2.5.3 Accept invitation
Now, Bob accepts the invitation.

````python
def accept_invitation():
    return send_request(f"{bob_endpoint}/didexchange/{bob_connection_id}/accept-invitation", {})

resp = accept_invitation()
print(f"Response: \n{pp_json(resp.json())}")
````

Note that the informations about whether the invitation was accepted is shown in the Bob's and Alice's logs. When Bob accepts the invitation, he sends a connection request to Alice, which happens automatically.

Even though Alice started this interaction by creating an invitation, it is actually Bob that sends a connection request to Alice, therefore Bob is called the requester. Alice is called the responder.

#### 2.5.4 Accept connection
Alice now accepts the connection request with her `connection_id`. Note that this ID is different for Alice and Bob. Alice must retrieve her `connection_id` by querying the `/connections` endpoint first.

````python
def get_connection_id():
    return send_request(alice_endpoint + "/connections", req_type="get")
resp = get_connection_id().json()

# Get alice's id
for el in resp["results"]:
    if "their_label" in el.keys() and el["their_label"] == "Bob" and el["rfc23_state"] == "request-received":
        alice_connection_id = el["connection_id"]
        break


print(f"Response: \n{pp_json(resp)}")
print(f"Alice's connection id: {alice_connection_id}")
````

*Alice* can now accept the connection request.

````python
def accept_connection():
    return send_request(f"{alice_endpoint}/didexchange/{alice_connection_id}/accept-request")

resp = accept_connection()
print(f"Response: {pp_json(resp.json())}")
````

The connection is established. 

#### 2.5.5 Test the connection

````python
def send_message(message, sender_endpoint, sender_conn_id):
    d = {
        "content" : message
    }
    return send_request(f"{sender_endpoint}/connections/{sender_conn_id}/send-message", d)

resp = send_message("This is Alice", alice_endpoint, alice_connection_id)
print(resp)
send_message("This is Bob", bob_endpoint, bob_connection_id)
print(resp)
````

### 2.6 Creating a schema and a credential definition
Alice will be issuing a credential to Bob. Before she can do that she needs to create a schema and a credential definition. The two attributes stored in a schema are *name* and *age*. By invoking the `/schemas` and `/credential-definitions` endpoints, Alice defines the schema and a credential definition, which are both written to the ledger automatically.  

````python
def create_schema():
    d = {
        "attributes": [
          "name",
          "age"
        ],
        "schema_name": "testtest",
        "schema_version": "1.0"
    }
    return send_request(f"{alice_endpoint}/schemas", d)

resp = create_schema()
schema_id = resp.json()["schema_id"]
print(f"Schema_id: {schema_id}")
print(f"Response: \n{pp_json(resp.json())}")
````

````python
def create_credential_definition(schema_id):
    d = {
        "schema_id": schema_id,
        "tag": "testtest"
    }
    return send_request(f"{alice_endpoint}/credential-definitions", d)

resp = create_credential_definition(schema_id)
print(f"Response: \n{pp_json(resp.json())}")
````

Now, a credential definition is created based upon the just created schema.

### 2.7 Issuing a credential

The credential issuing process can be executed in three different ways, based on who the initiator of the issuing process is:
1) The *holder* initializes the issuing process by sending a proposal to the *issuer*. This proposal contains information about what you would like to receive from the *issuer*. Based on that, the *issuer* sends an offer to the *holder*.

2) The *issuer* sends the offer to the *holder* directly, so no proposal is needed. 

3) The *holder* can directly send a request to the *issuer*, skipping the proposal and offer.

When the first option is used, the whole process consists of the next steps:
1) Holder sends a proposal to the issuer (issuer receives proposal)

2) Issuer sends an offer to the holder based on the proposal (holder receives offer)

3) Holder sends a request to the issuer (issuer receives request)

4) Issuer sends credential to holder (holder receives credentials)

5) Holder stores credential (holder sends acknowledge to issuer)

6) Issuer receives acknowledge

The process in action is shown in the figure [TODO ref].

````python
%%plantuml

@startuml
actor "Alice(issuer)" as A
actor "Bob(holder)" as B
database "Ledger" as L

A -> A : Create a schema and a credential definition
A -> L: Send credential definition to ledger

autonumber 
B -> A : Send proposal
A --> B : Send offer based on proposal
B -> A : Send request
A --> B : Send credential
B --> A : Store credential and send acknowledgement
B --> A : Receive acknowledge
@enduml
````

The whole process is demonstrated below.

#### 2.7.1 Bob sends a proposal to Alice
The `connection_id` is used.

````python
def send_proposal():
    d = {    
        "comment": "I need a credential. 1",
        "connection_id": bob_connection_id,
        "credential_preview": {
            "@type": "issue-credential/2.0/credential-preview",
            "attributes": [
                {
                "mime-type": "plain/text",
                "name": "name", 
                "value": "Bob"
                },
                {
                "mime-type": "plain/text",
                "name": "age", 
                "value": "30"
                }
            ]
        },
        "filter": {
            "dif": {},
            "indy": {}
        }
    }
    return send_request(f"{bob_endpoint}/issue-credential-2.0/send-proposal", d)

resp = send_proposal().json()
bob_cred_ex_id = resp["cred_ex_id"]

print(f"Response: \n{pp_json(resp)}")
print(f"cred_ex_id: {bob_cred_ex_id}")
````

The result is a Credential Exchange Record, which contains the state of the credential process. These records are stored in ACA-py and can be retrieved using the `/issue-credentials-2.0/records/{id}` endpoint where the `id` is the `cred_ex_id` in the result.

#### 2.7.2 Alice responds with an offer

````python
def get_alice_cred_id():
    resp = send_request(f"{alice_endpoint}/issue-credential-2.0/records", data={}, req_type="get").json()["results"]
    for el in resp:
        if el["cred_ex_record"]["cred_proposal"]["comment"] == "I need a credential. 1": # TODO: the corresponding proposal should be found differently
            return el["cred_ex_record"]["cred_ex_id"]

def respond_with_offer(alice_cred_id):
    return send_request(f"{alice_endpoint}/issue-credential-2.0/records/{alice_cred_id}/send-offer")

alice_cred_id = get_alice_cred_id()


resp = respond_with_offer(alice_cred_id)
print(pp_json(resp.json()))
        
````

#### 2.7.3 Bob requests the credential

````python
def request_credential():
    return send_request(f"{bob_endpoint}/issue-credential-2.0/records/{bob_cred_ex_id}/send-request", data={}, req_type="post")

resp = request_credential()
print(pp_json(resp.json()))
````

#### 2.7.4 Alice issues the credential

````python
def issue_credential():
    d = {"comment": "This is the credential."}
    return send_request(f"{alice_endpoint}/issue-credential-2.0/records/{alice_cred_id}/issue", data=d, req_type="post")

resp = issue_credential()
print(pp_json(resp.json()))
````

#### 2.7.4 Bob stores the received credential

````python
def store_credential():
    return send_request(f"{bob_endpoint}/issue-credential-2.0/records/{bob_cred_ex_id}/store")

resp = store_credential()
print(pp_json(resp.json()))
````

The holder can retrieve the stored credential by calling `/credentials` endpoint:

````python
def retrieve_credential():
    return send_request(f"{bob_endpoint}/credentials", req_type="get", data={})

resp = retrieve_credential()
print(pp_json(resp.json()))
````

### 2.8 Skipping the proposal
The *issuer* can directly send the offer to the *holder*, without the need for the proposal. This is done through the `/issue-credential-2.0/send-offer` endpoint.

````python
def send_offer():
    d = {
      "comment": "I can send you this credential",
      "connection_id": f"{alice_connection_id}", 
      "credential_preview": {
        "@type": "issue-credential/2.0/credential-preview", 
        "attributes": [
        {
          "name": "name", 
          "value": "Bob"
        },{
          "name": "age",
          "value": "30"
        }]
      },
      "filter": {
        "dif": {}, 
        "indy": {}
      }
    }
    return send_request(f"{alice_endpoint}/issue-credential-2.0/send-offer", data=d, req_type="post")
    
resp = send_offer()
print(pp_json(resp.json()))
````

After this offer, the flow continues with the holder responding with a request, as described above.
